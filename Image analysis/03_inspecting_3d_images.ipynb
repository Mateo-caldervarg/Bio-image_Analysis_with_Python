# Inspecting 3D image data
This notebook demonstrates how to navigate through 3D images.

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread
import stackview
```

```python
# Laod example data
mri_image = imread('Haase_MRT_tfl3d1.tif')
mri_image.shape
```

## Slicing
As also 3D images are just arrays, we can use the same syntax to select planes as if we were selecting array elements.

```python
#xy_plane = mri_image[100]
x = mri_image[1:4]
plt.imshow(x)
#plt.imshow(xy_plane)
```

```python
zx_plane = mri_image[:,128]

plt.imshow(zx_plane)
```

You can transpose axes using the `.T` transpose operation.

```python
plt.imshow(zx_plane.T)
```

## Projection
Projections along an axis allow to visualize 3D stacks as well. When specifying axes in Python, typically Z is the first axis (`0`), Y is the second (`1`) and X the last (`2` or `-1`)

```python
# Maximum Z Projection
projection = mri_image.max(axis=0)

# show result
plt.imshow(projection)
```

The average or "mean" projection appears more _transparent_:

```python
# Mean Z Projection
projection = mri_image.mean(axis=0)

# show result
plt.imshow(projection)
```

```python
# Maximum X Projection
projection = mri_image.max(axis=1)

# show result
plt.imshow(projection.T)
```

## Subplots

Use subplots to view multiple slices them side by side

```python
fig, axs = plt.subplots(1, 4, figsize=(15, 7))
axs[0].imshow(mri_image[:,75])
axs[1].imshow(mri_image[:,100])
axs[2].imshow(mri_image[:,125])
axs[3].imshow(mri_image[:,150])
```

## Interactive stack views
The [stackview]() library allows to browse the entire stack interactively. This will not be rendered on websites such as github.com. It only works by executing the notebook locally.

```python
import stackview

stackview.slice(mri_image)
```
